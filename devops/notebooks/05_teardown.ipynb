# 05 — Teardown

** **

This destroys all Chameleon resources:
- 3 VM instances (node1, node2, node3)
- Private network and subnet
- Router and floating IP
- 50GB block storage volume

> This is irreversible. All data on the cluster will be lost.


```bash
# runs in Chameleon Jupyter environment
export PATH=/work/.local/bin:$PATH

# Re-set your variables (same as in 01_provision.ipynb)
export TF_VAR_suffix=YOURNETID         # <-- replace
export TF_VAR_key=id_rsa_chameleon     # <-- replace if different
export TF_VAR_reservation=PASTE_UUID   # <-- paste reservation UUID from 01_provision.ipynb

# Unset OS_ vars
unset $(set | grep -o "^OS_[A-Za-z0-9_]*")

echo "Ready to destroy:"
echo "  suffix:      $TF_VAR_suffix"
echo "  key:         $TF_VAR_key"
echo "  reservation: $TF_VAR_reservation"
```

```bash
# runs in Chameleon Jupyter environment
export PATH=/work/.local/bin:$PATH
cd /work/immich-iac/tf/kvm
terraform destroy -auto-approve
```

After this completes, go to the KVM@TACC Horizon GUI and confirm the instances no longer appear.

**All resources destroyed.**
